```python
import pickle
import tensorflow as tf 
import matplotlib.pyplot as plt
import numpy as np

res = (28,28)
```

```python
#opening pickle file of augmented added data
with open('../data/external/combined_augmented_data_v2.pkl','rb') as f:
    X_train,y_train,X_test,y_test = pickle.load(f)
```

```python
model = tf.keras.models.load_model('../models/finalModel_V2.h5')
```

```python
#letters
letters = ['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y']
#numbers 
numbers = sorted(list(np.unique(y_train.astype(int))))
#dictionary of labels 
labels_dict = dict(zip(numbers,letters))
```

```python
def predict_with_nn(model,images,labels,index,res,labels_dict):
    img = images[index]
    plt.imshow(img)
    plt.show()
    print(f'True Label: {labels_dict[labels[index]]}')
    img = img.reshape((1,)+res+(1,))
    predictions = model.predict(img)
    print(f'predictions: {predictions}')
    index = np.argmax(predictions)
    letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'O', 'P', 'Q','R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y']
    print(f'result: {letters[index]}')


```

```python
for i in range(1,10):
    predict_with_nn(model,X_train,y_train,i,res,labels_dict)
```

```python
for i in range(1,10):
    predict_with_nn(model,X_test,y_test,i,res,labels_dict)
```
