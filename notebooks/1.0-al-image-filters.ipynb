```python
#import libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading image
img = cv.imread('../data/external/test.png')
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
#showing image
plt.imshow(img,cmap='gray')
```

```python
smooth = cv.GaussianBlur(img,(25,25),cv.BORDER_DEFAULT)
plt.imshow(smooth,cmap='gray')
```

```python
ret,threshold = cv.threshold(smooth,100,255,cv.THRESH_OTSU)
plt.imshow(threshold,cmap='gray')
```

```python
brighten = cv.convertScaleAbs(img,alpha=1,beta=80)
plt.imshow(brighten,cmap='gray')
```

```python
result = np.where(threshold==255,img,brighten)
plt.imshow(result,cmap='gray')
```

```python
division = cv.divide(img,smooth,scale=192)
plt.imshow(division,cmap='gray')
```

```python
thres = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                          cv.THRESH_BINARY, 73, 3)
plt.imshow(thres,cmap='gray')
```
